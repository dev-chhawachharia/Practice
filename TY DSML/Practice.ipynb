# Quesiton 4 and 23

```python
import math
```

```python
df = pd.read_csv("sales.csv")
df
eps = np.finfo(float).eps
```

```python
target = df.keys()[-1]
values = df[target].unique()
entropy = 0
for x in values:
    fraction = df[target].value_counts()[x]/len(df[target])
    entropy+= -(fraction*math.log(fraction,2))
```

```python

print(entropy)
```

```python
def cal_entropy(df,attribute):
    target = df.keys()[-1]
    values =  df[target].unique()
    variables = df[attribute].unique()
    entropy2 = 0
    for value in values:
        entropy=0
        for variable in variables:
            num = len(df[attribute][df[attribute]==variable][df[target]==value])
            den = len(df[attribute][df[attribute]==variable])
            fraction = (num)/den
            entropy += -fraction * math.log(fraction+eps,2)
        fraction2 = den/len(df)
        entropy2 += fraction2*entropy
    return entropy2
```

```python
IG = []
count=0
for i in df.keys()[:-1]:
    IG.append(entropy - cal_entropy(df,i))
    print(f"Coloumn Name: {i}\nInformation Gain:{IG[count]}")
    count+=1
root_node = df.keys()[:-1][np.argmax(IG)]
print("Root Node : ",root_node)
```

```python
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
iris_data = datasets.load_iris()
df = pd.DataFrame(iris_data.data , columns = iris_data.feature_names)
df = df.drop(df.tail(1).index)
df
```

```python
plt.scatter(df["petal length (cm)"],df["sepal length (cm)"])
```

```python
km = KMeans(n_clusters=3 , init="k-means++" , max_iter=10 , random_state=1)
y_kmeans=km.fit_predict(df)
```

```python
print(y_kmeans)
```

```python
cluster_1_population = np.count_nonzero(km.labels_==0)
print(cluster_1_population)
```

```python
cluster_2_population = np.count_nonzero(km.labels_==1)
print(cluster_2_population)
```

```python
cluster_3_population = np.count_nonzero(km.labels_==2)
print(cluster_3_population)
```
